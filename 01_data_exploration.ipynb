```python
# Clean price data - remove $ and convert to numeric
def clean_price(price_str):
    if pd.isna(price_str):
        return np.nan
    # Remove $ and commas, convert to float
    return float(re.sub(r'[\$,]', '', str(price_str)))

listings['price_numeric'] = listings['price'].apply(clean_price)

# Remove rows with missing coordinates or prices
df_clean = listings.dropna(subset=['latitude', 'longitude', 'price_numeric'])

# Remove extreme outliers (prices above 99th percentile or below 1st percentile)
price_lower = df_clean['price_numeric'].quantile(0.01)
price_upper = df_clean['price_numeric'].quantile(0.99)
df_clean = df_clean[(df_clean['price_numeric'] >= price_lower) & 
                   (df_clean['price_numeric'] <= price_upper)]

print(f"Data shape after cleaning: {df_clean.shape}")
print(f"Price range: ${df_clean['price_numeric'].min():.2f} - ${df_clean['price_numeric'].max():.2f}")

# Create multiple visualizations

# 1. Scatter plot with price color mapping
plt.figure(figsize=(15, 10))

# Create subplots
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(2, 2, figsize=(20, 16))

# Plot 1: Basic scatter plot
scatter1 = ax1.scatter(df_clean['longitude'], df_clean['latitude'], 
                      c=df_clean['price_numeric'], cmap='viridis', 
                      alpha=0.6, s=20)
ax1.set_xlabel('Longitude')
ax1.set_ylabel('Latitude')
ax1.set_title('Listing Prices by Geographic Location')
plt.colorbar(scatter1, ax=ax1, label='Price ($)')

# Plot 2: Hexbin plot for density
hb = ax2.hexbin(df_clean['longitude'], df_clean['latitude'], 
                C=df_clean['price_numeric'], gridsize=30, cmap='Reds', 
                reduce_C_function=np.mean)
ax2.set_xlabel('Longitude')
ax2.set_ylabel('Latitude')
ax2.set_title('Average Price Density (Hexagonal Binning)')
plt.colorbar(hb, ax=ax2, label='Average Price ($)')

# Plot 3: Price categories
# Create price categories
df_clean['price_category'] = pd.cut(df_clean['price_numeric'], 
                                   bins=5, labels=['Very Low', 'Low', 'Medium', 'High', 'Very High'])

colors = ['green', 'lightgreen', 'yellow', 'orange', 'red']
for i, category in enumerate(['Very Low', 'Low', 'Medium', 'High', 'Very High']):
    mask = df_clean['price_category'] == category
    ax3.scatter(df_clean[mask]['longitude'], df_clean[mask]['latitude'], 
               c=colors[i], label=category, alpha=0.6, s=15)

ax3.set_xlabel('Longitude')
ax3.set_ylabel('Latitude')
ax3.set_title('Price Categories by Location')
ax3.legend()

# Plot 4: Price distribution histogram
ax4.hist(df_clean['price_numeric'], bins=50, alpha=0.7, color='skyblue', edgecolor='black')
ax4.set_xlabel('Price ($)')
ax4.set_ylabel('Frequency')
ax4.set_title('Price Distribution')
ax4.axvline(df_clean['price_numeric'].mean(), color='red', linestyle='--', 
           label=f'Mean: ${df_clean["price_numeric"].mean():.2f}')
ax4.axvline(df_clean['price_numeric'].median(), color='orange', linestyle='--', 
           label=f'Median: ${df_clean["price_numeric"].median():.2f}')
ax4.legend()

plt.tight_layout()
plt.show()

# Print summary statistics
print("\nPrice Statistics by Neighbourhood:")
print("="*50)
avg_price_by_neighborhood = df_clean.groupby('neighbourhood_cleansed')['price_numeric'].mean().sort_values(ascending=False)
print(avg_price_by_neighborhood)
```

```python
# --- Columns for Scatter Plots ---
scatter_cols = [
    'host_response_rate', 'host_acceptance_rate', 'host_is_superhost',
    'host_listings_count', 'host_total_listings_count', 'host_has_profile_pic',
    'host_identity_verified', 'bathrooms', 'bedrooms', 'beds',
    'minimum_nights', 'maximum_nights', 'minimum_minimum_nights',
    'maximum_minimum_nights', 'minimum_maximum_nights', 'maximum_maximum_nights',
    'minimum_nights_avg_ntm', 'maximum_nights_avg_ntm',
    'availability_30', 'availability_60', 'availability_90', 'availability_365', 'availability_eoy',
    'number_of_reviews', 'number_of_reviews_ltm', 'number_of_reviews_l30d',
    'number_of_reviews_ly', 'review_scores_rating', 'review_scores_accuracy',
    'review_scores_cleanliness', 'review_scores_checkin', 'review_scores_communication',
    'review_scores_location', 'review_scores_value', 'instant_bookable',
    'calculated_host_listings_count', 'calculated_host_listings_count_entire_homes',
    'calculated_host_listings_count_private_rooms', 'calculated_host_listings_count_shared_rooms',
    'reviews_per_month'
]

# Create a copy to avoid SettingWithCopyWarning
df_scatter = df_clean.copy()


# --- Data Cleaning & Transformation ---

# Correctly handle percentage columns that may not be strings
def convert_percent_to_float(value):
    if isinstance(value, str):
        try:
            return float(value.strip().replace('%', '')) / 100
        except (ValueError, AttributeError):
            return np.nan
    return value

for col in ['host_response_rate', 'host_acceptance_rate']:
    df_scatter[col] = df_scatter[col].apply(convert_percent_to_float)


# Convert 't'/'f' columns to 1/0
t_f_cols = ['host_is_superhost', 'host_has_profile_pic', 'host_identity_verified', 'instant_bookable']
for col in t_f_cols:
    df_scatter[col] = df_scatter[col].map({'t': 1, 'f': 0})


# --- Plotting ---

# Determine the grid size for subplots
num_plots = len(scatter_cols)
num_cols = 4  # Number of plots per row
num_rows = (num_plots + num_cols - 1) // num_cols # Calculate required number of rows

# Create the figure and axes for the subplots
fig, axes = plt.subplots(num_rows, num_cols, figsize=(20, num_rows * 5))
axes = axes.flatten() # Flatten to a 1D array for easy iteration

# Loop through the columns and create a scatter plot for each
for i, col in enumerate(scatter_cols):
    ax = axes[i]
    
    # Drop missing values only for the columns in the current plot
    plot_data = df_scatter[[col, 'price_numeric']].dropna()
    
    # Use a low alpha to see density of points
    ax.scatter(plot_data[col], plot_data['price_numeric'], alpha=0.2, s=15)
    ax.set_xlabel(col, fontsize=12)
    ax.set_ylabel('Price ($)', fontsize=12)
    ax.set_title(f'Price vs. {col}', fontsize=14)

# Hide any empty, unused subplots at the end of the grid
for i in range(num_plots, len(axes)):
    axes[i].set_visible(False)

# Adjust layout to prevent titles and labels from overlapping
plt.tight_layout()
plt.show()
```

```python
# --- Correlation Analysis ---

# Select all the numeric columns we've prepared for the correlation matrix
numeric_cols_for_corr = [
    'host_response_rate', 'host_acceptance_rate', 'host_is_superhost',
    'host_listings_count', 'host_total_listings_count', 'host_has_profile_pic',
    'host_identity_verified', 'bathrooms', 'bedrooms', 'beds',
    'minimum_nights', 'maximum_nights', 'minimum_minimum_nights',
    'maximum_minimum_nights', 'minimum_maximum_nights', 'maximum_maximum_nights',
    'minimum_nights_avg_ntm', 'maximum_nights_avg_ntm',
    'availability_30', 'availability_60', 'availability_90', 'availability_365', 'availability_eoy',
    'number_of_reviews', 'number_of_reviews_ltm', 'number_of_reviews_l30d',
    'number_of_reviews_ly', 'review_scores_rating', 'review_scores_accuracy',
    'review_scores_cleanliness', 'review_scores_checkin', 'review_scores_communication',
    'review_scores_location', 'review_scores_value', 'instant_bookable',
    'calculated_host_listings_count', 'calculated_host_listings_count_entire_homes',
    'calculated_host_listings_count_private_rooms', 'calculated_host_listings_count_shared_rooms',
    'reviews_per_month', 'price_numeric'
]

# Create a new DataFrame with just these numeric columns
df_corr = df_scatter[numeric_cols_for_corr].copy()

# Calculate the correlation matrix
correlation_matrix = df_corr.corr()

# --- Visualization 1: Correlation Heatmap ---

# Create a heatmap to visualize the correlation matrix
plt.figure(figsize=(20, 18))
sns.heatmap(correlation_matrix, annot=False, cmap='coolwarm', linewidths=.5)
plt.title('Correlation Matrix of All Variables', fontsize=16)
plt.show()


# --- Visualization 2: Bar Chart of Price Correlations ---

# Get the correlation of all variables with 'price_numeric'
price_correlation = correlation_matrix['price_numeric'].drop('price_numeric').sort_values(ascending=False)

# Create a bar chart
plt.figure(figsize=(12, 14))
sns.barplot(x=price_correlation.values, y=price_correlation.index, palette='vlag')
plt.title('Correlation of Variables with Price', fontsize=16)
plt.xlabel('Correlation Coefficient', fontsize=12)
plt.ylabel('Variables', fontsize=12)
plt.axvline(x=0, color='grey', linewidth=0.8) # Add a line at 0 for reference
plt.show()
```

```python
import ast
from collections import Counter
# We use a try-except block to handle any potential errors if a value is not a valid list
def parse_amenities(amenities_str):
    try:
        return ast.literal_eval(amenities_str)
    except (ValueError, SyntaxError):
        return []

df_amenities = df_clean.copy()
df_amenities['amenities_list'] = df_amenities['amenities'].apply(parse_amenities)

# --- 1. Count the frequency of each amenity ---

# Create a single list of all amenities from all listings
all_amenities = [amenity for sublist in df_amenities['amenities_list'] for amenity in sublist]

# Count the occurrences of each amenity
amenity_counts = Counter(all_amenities)

# Get the 20 most common amenities
top_amenities = amenity_counts.most_common(20)
top_amenity_names = [amenity[0] for amenity in top_amenities]

# --- 2. Calculate average price for each top amenity ---

# Create a dictionary to store the average price for each amenity
avg_price_per_amenity = {}

for amenity in top_amenity_names:
    # Create a boolean mask for rows containing the current amenity
    has_amenity_mask = df_amenities['amenities_list'].apply(lambda x: amenity in x)
    
    # Calculate the mean price for listings that have the amenity
    avg_price = df_amenities[has_amenity_mask]['price_numeric'].mean()
    avg_price_per_amenity[amenity] = avg_price

# Convert the dictionary to a pandas Series for easy sorting and plotting
avg_price_series = pd.Series(avg_price_per_amenity).sort_values(ascending=False)


# --- 3. Visualization ---

plt.figure(figsize=(12, 10))
sns.barplot(x=avg_price_series.values, y=avg_price_series.index, palette='crest')

plt.title('Average Price by Top 20 Amenities', fontsize=16)
plt.xlabel('Average Price ($)', fontsize=12)
plt.ylabel('Amenity', fontsize=12)
plt.show()
```
